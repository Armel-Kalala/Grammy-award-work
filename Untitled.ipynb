```python
# importation des bibliothèque
import pandas as pd
```

```python
# téléchargements du dataset
data = pd.read_csv('Grammy Award Nominees and Winners 1958-2024.csv')
```

```python
# affichage du dataset
data
```

```python
# renomer les colonnes en français
data.rename(columns={
    'Year':'Annee', 
    'Ceremony':'Ceromonie',
    'Award ID':'Prix ID',
    'Award Type':'Type_prix',
    'Award Name':'Nom_prix',
    'Work':'Travail',
    'Nominee':'Candidate',
    'Winner':'Gagnante',
},inplace=True)
```

```python
# suppression de la colonne d'index inutile si présente
if 'Unnamed: 0' in data.columns:
    data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# combien de gagnats par anneé et prix
winners_per_award=data[data['Gagnante']==True].groupby(['Annee', 'Nom_prix']).size()
multiple_winners=winners_per_award[winners_per_award>1]
multiple_winners
```

```python
# supprimer les lignes trop vides 
data.dropna(subset=['Nom_prix', 'Candidate'],)
```

```python
# remplacer les NAN dans la colonne travail
data['Travail'].fillna('Inconnu',inplace=True)
```
